Consumo de API do IBGE 

Retorno esperado : Dados sobre Pesquisa Mensal do Comércio

```python
import pandas as pd  #possibilita manipulação e análise de dados
import requests  #possibilita integração com serviços web, solicitações HTTP consumir dados de APIs
```

Índice de volume de vendas no comércio varejista

```python
url = "https://servicodados.ibge.gov.br/api/v3/agregados/8880/periodos/201608|201609|201610|201611|201612|201701|201702|201703|201704|201705|201706|201707|201708/variaveis/11710?localidades=N1[all]&classificacao=11046[56734]"
```

```python
# Realizar a requisição
response = requests.get(url)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    # Converter os dados para JSON
    data = response.json()
```

Abaixo busca o valor do índice PMC, que é : Variação acumulada no ano (em relação ao mesmo período do ano anterior)

```python
# Extração e exibição dos valores de interesse
try:
        resultados = data[0]["resultados"][0]["series"]  # Navega até os dados
        for serie in resultados:
            localidade = serie["localidade"]["nome"]  # Nome da localidade
            print(f"Localidade: {localidade}")
            
            for periodo, valor in serie["serie"].items():  # Itera sobre os períodos e valores
                print(f"Período: {periodo}, Valor do Indice: {valor}")
except KeyError as e:
        print(f"Erro ao processar os dados: {e}")
```
