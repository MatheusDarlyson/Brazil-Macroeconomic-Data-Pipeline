Consumo de API do IBGE 

Retorno esperado : Dados sobre a Inflação IPCA

```python
import pandas as pd  #possibilita manipulação e análise de dados
import requests  #possibilita integração com serviços web, solicitações HTTP consumir dados de APIs
```

```python
url = "https://servicodados.ibge.gov.br/api/v3/agregados/1419/periodos/-6/variaveis/63?localidades=N1[all]&classificacao=315[7736,7792]"
```

```python
# Realizar a requisição
response = requests.get(url)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    # Converter os dados para JSON
    data = response.json()
```

```python
# Processar os dados para criar uma tabela
tabela_dados = []

for resultado in data[0]["resultados"]:
    categoria = list(resultado["classificacoes"][0]["categoria"].values())[0]
    for serie in resultado["series"]:
        localidade = serie["localidade"]["nome"]
        for periodo, valor in serie["serie"].items():
            tabela_dados.append({
                "Categoria": categoria,
                "Localidade": localidade,
                "Período": periodo,
                "Valor (%)": float(valor)
            })

# Criar DataFrame
df = pd.DataFrame(tabela_dados)

# Transformar o Período para um formato legível (ano e mês)
df["Período"] = pd.to_datetime(df["Período"], format="%Y%m")

# Exibir a tabela no Jupyter Notebook
from IPython.display import display
display(df)
```
