```python
import pandas as pd  #possibilita manipulação e análise de dados
import requests  #possibilita integração com serviços web, solicitações HTTP consumir dados de APIs
from IPython.display import display
```

```python
url = "https://servicodados.ibge.gov.br/api/v3/agregados/1628/periodos/201510/variaveis/708|19|21|20|22|685|686?localidades=N1[all]&classificacao=24[2542]|11773[95038]"
```

```python
# Realizar a requisição
response = requests.get(url)
```

```python
# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    # Converter os dados para JSON
    data = response.json()

    # Extração e organização dos valores em formato de tabela
    tabela_dados = []

    try:
        # Acesse os dados conforme a estrutura JSON retornada
        for resultado in data[0]["resultados"]:
            for serie in resultado["series"]:
                localidade = serie["localidade"]["nome"]
                for periodo, valor in serie["serie"].items():
                    tabela_dados.append({
                        "Localidade": localidade,
                        "Período": periodo,
                        "Valor": valor
                    })

        # Criar DataFrame com os dados extraídos
        df = pd.DataFrame(tabela_dados)

       # Exibir a tabela diretamente no Jupyter Notebook
        display(df)

    except KeyError as e:
        print(f"Erro ao processar os dados: {e}")
else:
    print(f"Erro na requisição: {response.status_code}")
```
