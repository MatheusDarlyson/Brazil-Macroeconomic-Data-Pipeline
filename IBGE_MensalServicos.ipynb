Consumo de API do IBGE 

Retorno esperado : Dados sobre Pesquisa Mensal de Serviços

```python
import pandas as pd  #possibilita manipulação e análise de dados
import requests  #possibilita integração com serviços web, solicitações HTTP consumir dados de APIs
```

Resultados das atividades de transportes e serviços auxiliares dos transportes

```python
url = "https://servicodados.ibge.gov.br/api/v3/agregados/1526/periodos/2001/variaveis/643?localidades=N1[all]&classificacao=11571[93041]"
```

Receita operacional líquida de empresas de transporte rodoviários de passageiros em 2001

```python
# Realizar a requisição
response = requests.get(url)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    # Converter os dados para JSON
    data = response.json()
```

```python
# Extração e exibição dos valores de interesse
try:
        resultados = data[0]["resultados"][0]["series"]  # Navega até os dados
        for serie in resultados:
            localidade = serie["localidade"]["nome"]  # Nome da localidade
            print(f"Localidade: {localidade}")
            
            for periodo, valor in serie["serie"].items():  # Itera sobre os períodos e valores
                print(f"Período: {periodo}, Receita: {valor}")
except KeyError as e:
        print(f"Erro ao processar os dados: {e}")
```
